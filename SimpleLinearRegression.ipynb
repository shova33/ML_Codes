```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML_Codes/Linear_Regression/placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('cgpa')
plt.ylabel('package')
plt.title('scatter plot')
plt.show()
```

```python
X = df.iloc[:,0:1]
y= df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
lr.predict(X_test.iloc[2].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_test,lr.predict(X_test),color='red')
plt.xlabel('cgpa')
plt.ylabel('package')
plt.title('scatter plot')
plt.show()
```

```python
m = lr.coef_
m
```

```python
c = lr.intercept_
c
```

```python
m * 9.5 + c
```
